```python
#Import Dependencies

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot w/ MATLAB

plt.plot(x_axis, y_axis)
```

```python
#Create the plots with ax.plt() aka OOI method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend.
plt.legend()
ax.plot(x_axis, y_axis, label = 'line')
```

# Scatter Plots and Bubble Charts

```python
# there are 2 ways to create scatter plots with MATLAB approach -- plt.plot() or plt.scatter()

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
#When we use the plt.plot() function to create a scatter plot, we need to add a lowercase "o" as a parameter inside the parentheses. This switches the plot from a line chart to a scatter plot.

plt.plot(x_axis, y_axis, 'o')
```

```python
# use of plt.scatter
plt.scatter(x_axis, y_axis)
```

```python
#SKILL DRILL
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.xlim(0,50)
```

```python
#Create a bubble chart
#This allows us to add a third factor -- size of the buble. We can add a fourth factor through color

plt.scatter(x_axis, y_axis)
```

```python
#Change the size of marker on the graph using "s" -- will change marker size based on the value of the fare USD
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
#multiply each data point on y axis by 3 to adjust size
#We do this by iterating through y axis data and multiply each data point by 3 and add it to a new list

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refactor code to use Python list comprehension technique instead of for loop.
# Format for list comprehension: new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```
